```python
# Finding Null Values If Any

df.isnull().sum()
```

```python
# Inputing Missing Values in Review Ranking Column with median rating of product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# converting columns according to snake case for better readability abd documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group

labels = ['Young-Adult', 'Adult', 'Middle-Aged', 'Senior']

df['age-group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age-group']].head(10)
```

```python
# create a new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']]
```

```python
# top 10 rows from both columns

df[['discount_applied','promo_code_used']].head(10)
```

```python
# comparing common elements from both columns 

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# dropping promo_code column

df = df.drop('promo_code_used', axis = 1)
```
